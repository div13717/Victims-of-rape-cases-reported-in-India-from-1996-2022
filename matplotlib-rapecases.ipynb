```python
plt.figure(figsize=(40, 15))

# to show dot at peak point

peak_y = max(casesreported)           # Maximum y value
peak_index =casesreported.index(peak_y)  # Index of the peak point
peak_x = Year[peak_index]   
plt.plot(Year,casesreported, linewidth=5)
plt.plot(peak_x, peak_y, marker='o', markersize=25, color="black", label="Peak Point")




# to highlight selected linesegment

start_index = 18  
end_index = 19  
x_selected = Year[start_index:end_index + 1]
y_selected = casesreported[start_index:end_index + 1]
plt.plot(x_selected, y_selected, marker='o', label="Selected Segment", color="black", linewidth=10)

start_index = 15  
end_index = 16  
x_selected = Year[start_index:end_index + 1]
y_selected = casesreported[start_index:end_index + 1]
plt.plot(x_selected, y_selected, marker='o', label="Selected Segment", color="black", linewidth=10)




# to show text at mid point of selected line segment 

mid_x = (Year[start_index] + Year[end_index]) / 2
mid_y = (casesreported[start_index] + casesreported[end_index]) / 2
plt.text(mid_x + 0.2, mid_y, " 35.52% increase", fontsize=50, color="red")

start_index = 15  # Index of the first point
end_index = 16    # Index of the second point




# to calculate percetage increase between selected points 

y1 = casesreported[start_index]  # Old Value
y2 = casesreported[end_index]    # New Value

# Calculate percentage increase
percentage_increase = ((y2 - y1) / y1) * 100

# Print the result
print()
print()
print(f"Percentage Increase between ({Year[start_index]}, {y1}) and ({Year[end_index]}, {y2}): {percentage_increase:.2f}%")
print()
print()




plt.xticks(ticks=Year, labels=Year)

plt.xticks(fontsize=29,fontweight='bold')
plt.yticks(fontsize=35,fontweight='bold')

plt.xlabel('Victims of rape cases',fontsize=37,labelpad=26,fontweight='bold')
plt.ylabel('Years',fontsize=37,labelpad=26,fontweight='bold')

plt.tight_layout()
plt.show()
```
